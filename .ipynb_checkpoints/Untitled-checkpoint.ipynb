```python
#---- EDA phase ----
#In this phase we will use the previously created csv of amazon option contracts.
#First we must open this csv and check that everything is right.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

options = pd.read_csv("amazon_options_clean.csv")
display(options.head())

print("We will first make a volume analisis, to understand what contracts are more popular and why.")
display(options["volume"].describe())
print("We can first see that most contracts (75% lowest percentile) dont have a lot of volume. We will take a look at the top 10% contracts by volume. We can see the difference in the following graphs:")

cut = options["volume"].quantile(0.90)
top_percent = options[options["volume"] >= cut].copy()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].hist(options['volume'], bins=50, edgecolor='black')
axes[0].set_title('Volume Distribution - All Contracts')
axes[1].hist(top_percent['volume'], bins=30, edgecolor='black')
axes[1].set_title('Volume Distribution - Top 10%')
plt.tight_layout()
plt.show()

or_len = len(options)
print(f"We have dropped {or_len - len(top_percent)} contracts with low volume.")
display(top_percent)

print("\nNow we will investigate these popular options contracts to see if we can find a pattern and a reason for their higher volume.")
print("That's why we will ask the following questions:\n-->Are the contracts mostly calls or puts?\n-->Are they far or near expiration?\n-->Regarding moneyness, are they ITM, ATM or OTM?")
print("-->What strike prices are more popular in relation to the current price of the stock?")
```

```python
print("Contract type distribution:\n")
print(options["contract_type"].value_counts(), "\n")
calls = (options["contract_type"] == "Calls").sum()
puts = (options["contract_type"] == "Puts").sum()
if calls > puts:
    if puts + 100 > calls:
        print("The number of call and put option contracts is similar. The difference is:", calls - puts)
        print("We can assume that there is no direct relation between the volume and the type of the contract. The sentiment on the future price of the stock is balanced.")
    else:
        print("The number of calls is significantly higher than the number of put contracts. The difference is:", calls - puts)
if puts > calls:
    if calls + 100 > puts:
        print("The number of call and put option contracts is similar. The differnce is:", puts - calls)
        print("We can assume that there is no direct relation between the volume and the type of the contract. The sentiment on the future price of the stock is balanced.")
    else:
        print("The number of puts is significantly higher than the number of call contracts. The difference is:", puts - calls)
print("We will now take a look at the relation between the days until expiration and the volume. For this we will also look at at the whole dataset and the 10% top percetile dataset.")
print("\nAverage days to expiration (high volume):", top_percent['days_to_exp'].mean())
print(f"Average days to expiration (all):", options['days_to_exp'].mean())


```
